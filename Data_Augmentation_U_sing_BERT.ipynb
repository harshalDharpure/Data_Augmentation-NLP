```python
#Creat a masked word Prediction Pipeline
```

```python
from transformers import pipeline

def create_pipeline(model_name):
   return pipeline('fill-mask', model=model_name)

predict_mask = create_pipeline("bert-base-uncased")
```

```python
predict_mask("I am Good at [MASK].")
```

```python
# Using Multilingual BERT (Bidirectional Encoder Representation From Transformers)
```

```python
predict_mask_ml=create_pipeline("bert-base-multilingual-uncased")
```

```python
predict_mask_ml("Berlin hat den [MASK] bahnhof in europa.")
```
